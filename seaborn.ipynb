```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
#catplot
sns.catplot(x="total_bill", y = "tip", data = df,kind= "strip")
```

```python
sns.catplot(x = "day",y = "total_bill", data = df,kind = "box")
```

```python
sns.catplot(x = "day",y = "total_bill", data = df,kind = "swarm")
```

```python
#countplot
sns.countplot(x = "day",data = df)
```

```python
##pointplot---shows the mean of the data connected by lines to show the
#trend 
sns.pointplot(x = "day", y= "total_bill",data = df,hue = "sex")
```

```python
##boxplot 
sns.boxplot(x = "day",y = "total_bill",data =df)
```

```python
##violinplot---spread of data in the 
sns.violinplot(x = "day",y = "total_bill",data = df)
```

```python
sns.pairplot(df)
```

```python
sns.histplot(df["tip"],kde=True)
```

```python
sns.histplot(df["tip"],kde=False)
```

```python
sns.boxplot(df["tip"])
```

```python
##outliers
q1 = df["total_bill"].quantile(0.25)
q3 = df["total_bill"].quantile(0.75)
iqr = q3 - q1
lower_limit = q1-1.5*iqr
upper_limit=  q3+1.5*iqr
print(lower_limit,upper_limit)
df[df["total_bill"]>upper_limit]
df = df[(df["total_bill"]>lower_limit) & (df["total_bill"]<upper_limit)]
sns.boxplot(df["total_bill"])
```
